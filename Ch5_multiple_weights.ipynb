# Корректировка множества весов

```python
def neural_network(input, weights):
    """a neural network with multiple inputs

    Args:
        input (list): list of input values
        weights (list): list of weights

    Returns:
        float: output value
    """    
    out = 0
    for i in range(len(input)):
        out += (input[i] * weights[i])
    return out

def ele_mul(number, vector):
    """element-wise multiplication

    Args:
        number (float): number to multiply
        vector (list): list of numbers

    Returns:
        list: list of numbers
    """    
    output = [0, 0, 0]
    assert(len(output) == len(vector))
    for i in range(len(vector)):
        output[i] = number * vector[i]
    return output

weights = [2, -.8, -1]
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1, 1, 0, 1]
true = win_or_lose_binary[0]

alpha = 0.01
input = [toes[0], wlrec[0], nfans[0]]

iterations = {}
iterations["weights"] = []
iterations["error"] = []

for iter in range(30):
    pred = neural_network(input, weights)

    error = (pred - true) ** 2
    deltas = pred - true

    weight_deltas = ele_mul(deltas, input)

    print("Iteration: " + str(iter + 1))
    print("Pred: " + str(pred))
    print("Error: " + str(error))
    print("Delta: " + str(deltas))
    print("Weights: " + str(weights))
    print("Weight_Deltas: ")
    print(str(weight_deltas))
    print()
    iterations["weights"].append(weights.copy())
    iterations["error"].append(error)
    for i in range(len(weights)):
        weights[i] -= alpha * weight_deltas[i]
    


    
# plot the error and weights over time
import matplotlib.pyplot as plt
plt.plot(iterations["error"])
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.show()

# plot the weights over time as three subplots
fig, axs = plt.subplots(3)
fig.suptitle('Weights over time')
axs[0].plot([w[0] for w in iterations["weights"]])
axs[0].set_ylabel('Weight 0')
axs[1].plot([w[1] for w in iterations["weights"]])
axs[1].set_ylabel('Weight 1')
axs[2].plot([w[2] for w in iterations["weights"]])
axs[2].set_ylabel('Weight 2')
plt.show()


```

|-|-|-|-|

|0.1|0.2|0.0|победа|
|0.0|1.3|0.1|печаль|

```python
def w_sum(a, b):
    """
    a weighted sum of two numbers
    """
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output
```

```python
# Test case 1: Equal length lists
a = [1, 2, 3]
b = [4, 5, 6]
expected_output = 32
assert w_sum(a, b) == expected_output

# Test case 2: Empty lists
a = []
b = []
expected_output = 0
assert w_sum(a, b) == expected_output

# Test case 3: Lists with negative numbers
a = [-1, -2, -3]
b = [4, -5, 6]
expected_output = -12
assert w_sum(a, b) == expected_output

# Test case 4: Lists with floating-point numbers
a = [1.5, 2.5, 3.5]
b = [4.5, 5.5, 6.5]
expected_output = 43.25
assert w_sum(a, b) == expected_output

print("All test cases passed!")
```
